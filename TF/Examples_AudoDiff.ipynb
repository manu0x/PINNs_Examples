```python
### Example of how to do autograd in TF
# Here we see how we can take gradients of function generating a certain polynomial
```

```python
# The polynomial whose derivative we want to take

def  poly(x):
    f = 5.0*x**3+3.0*x+4.0
   
    return f


# Exact polynomial gradient for comparison
def dx_poly(x):
    df = 15.0*x*x+3.0
  
    return df
 
# Exact polynomial (2nd)gradient for comparison
def dx2_poly(x):
    ddf = 30.0*x
    
    return ddf






# using TF's autograd to compute derivatives of function "poly(x)" defined above    
@tf.function
def grad_poly(x):
    #This context records all operations so that the derivatives can be taken
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(x)
        f = poly(x)   # Calling the function whose derivatives we want, this could be any piece of code 
                      # that takes some input and calculates the output using well-defined maths operations(i.e.
                      # all intermediate operations have defined derivative), an example being Neural nets.

        dx_f = tape.gradient(f,x)
                                    # First derivative # Since we want to get 2nd derivatives also, we also record 
                                   # the first derivative operation on tape
            
    dxx_f = tape.gradient(dx_f,x)  # Second derivative

    del tape
    
    return dx_f,dxx_f
```

```python
x= tf.random.uniform((100,1),minval=0.0,maxval=1.0)


f = poly(x)
dx_f = dx_poly(x)
dxx_f = dx2_poly(x)

print(f.shape,dx_f.shape,dxx_f.shape)

#tf.summary.trace_on(graph=True, profiler=True) 


grd_x_f,grd_xx_f = grad_poly(x)   

#with writer.as_default():                    
#    tf.summary.trace_export(
#      name="One_comp_func_trace",
#      step=0,
 #     profiler_outdir=logdir)
    
print(grd_x_f.shape,grd_xx_f.shape)
plt.plot(x,grd_x_f.numpy(),"ro",label="autograd ")
plt.plot(x,dx_f.numpy(),"b+",label="exact")


plt.legend()
```

```python
########## Derivatives of a function from R->RxR i.e. one-D intput and has ouput with components
```

```python
# The polynomial whose derivative we want to take
def  poly(x):
    f1 = 5.0*x**3+3.0*x+4.0
    f2 = x**2+3.0*x+4.0
    return tf.concat((f1,f2),axis=1)


# Exact polynomial gradient for comparison
def dx_poly(x):
    f1 = 15.0*x*x+3.0
    f2 = 2.0*x+3.0
    return tf.concat((f1,f2),axis=1)
 
# Exact polynomial (2nd)gradient for comparison
def dx2_poly(x):
    f1 = 30.0*x
    f2 = 2.0*tf.ones_like(f1)
    return tf.concat((f1,f2),axis=1)

x= tf.random.uniform((100,1),minval=0.0,maxval=1.0)


f = poly(x)
dx_f = dx_poly(x)
dxx_f = dx2_poly(x)

print(f.shape,dx_f.shape,dxx_f.shape)




# using TF's autograd to compute derivatives of function "poly(x)" defined above    
@tf.function
def grad_poly(x):
    #This context records all operations so that the derivatives can be taken
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(x)
        f = poly(x)   # Calling the function whose derivatives we want, this could be any piece of code 
                      # that takes some input and calculates the output using well-defined maths operations(i.e.
                      # all intermediate operations have defined derivative), an example being Neural nets.
         
        
        f1 = f[:,0]
        f2 = f[:,1]
        dx_f1 = tape.gradient(f1,x)
        dx_f2 = tape.gradient(f2,x)# First derivative # Since we want to get 2nd derivatives also, we also record 
                                   # the first derivative operation on tape
            
    dxx_f1 = tape.gradient(dx_f1,x)  # Second derivative
    dxx_f2 = tape.gradient(dx_f2,x)
    del tape
    
    return tf.concat((dx_f1,dx_f2),axis=1),tf.concat((dxx_f1,dxx_f2),axis=1)
```

```python
x= tf.random.uniform((100,1),minval=0.0,maxval=1.0)


f = poly(x)
dx_f = dx_poly(x)
dxx_f = dx2_poly(x)

print(f.shape,dx_f.shape,dxx_f.shape)


grd_x_f,grd_xx_f = grad_poly(x)

print(grd_x_f.shape,grd_xx_f.shape)
plt.plot(x,grd_x_f[:,0].numpy(),"ro",label="autograd comp1")
plt.plot(x,dx_f[:,0].numpy(),"b+",label="exact comp1")

plt.plot(x,grd_x_f[:,1].numpy(),"go",label="autograd comp2")
plt.plot(x,dx_f[:,1].numpy(),"k+",label="exact comp2")
plt.legend()
```
