```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras as keras
from time import time
from PINNs import PINN
```

```python
DTYPE='float32'
tf.keras.backend.set_floatx(DTYPE)
pi = tf.constant(np.pi,dtype=DTYPE)
```

```python
class swe(PINN):
    def __init__(self,dom_bounds,out_dim,prob_coef,DTYPE='float32'):
        super().__init__(dom_bounds,out_dim,prob_coef,DTYPE)
    
    @tf.function
    def diff_eqn(self,dom_points):
        t,x,y = dom_points[:,0:1],dom_points[:,1:2],dom_points[:,2:3]
        
        with tf.GradientTape(persistent=True) as tape:
            tape.watch(t)
            tape.watch(x)
            tape.watch(y)
        
            f = self.model(tf.stack((t[:,0],x[:,0],y[:,0]),axis=1),training=True)
            
            h = f[:,0:1]
            u = f[:,1:2]
            v = f[:,2:3]
            
            hu = h*u
            hv = h*v
            huv = h*v*u
            
            mflx = h*tf.square(u)+0.5*self.prob_coef[0]*tf.square(h)
            mfly = h*tf.square(v)+0.5*self.prob_coef[0]*tf.square(h)
            
        
            
        
        dt_h = tape.gradient(h,t)
        dt_hu = tape.gradient(hu,t)
        dt_hv = tape.gradient(hv,t)
        
        dx_hu = tape.gradient(hu,x)
        dy_hv = tape.gradient(hv,y)
        
        dx_huv = tape.gradient(huv,x)
        dy_huv = tape.gradient(huv,y)
        
        
        
        
        
        dx_mflx = tape.gradient(mflx,x)
        dy_mfly = tape.gradient(mfly,y)
        
    
        del tape
        eqn1 = dt_h+dx_hu+dy_hv
        eqn2 = dt_hu + dx_mflx + dy_huv
        eqn3 = dt_hv + dy_mfly + dx_huv
        #print(eqn1.shape,eqn2.shape)
        return tf.concat([eqn1,eqn2,eqn3],axis=1) 
```

```python
rInflow = 0.1
rOutflow = 5.0
def get_domain_pointsInr_lhs(r_min=rInflow,r_max=rOutflow,t_min=0.0,t_max=3.0,n=100):
    
    from scipy.stats import qmc
    sampler = qmc.LatinHypercube(d=3)
    sample = sampler.random(n)
    l_bounds = [t_min,r_min,0.0]
    u_bounds = [t_max,r_max,2.0*pi]
    sample_scaled = qmc.scale(sample, l_bounds, u_bounds)
    t = tf.constant(1.0*sample_scaled[:,0],dtype=DTYPE)
    r = tf.constant(1.0*sample_scaled[:,1],dtype=DTYPE)
    theta =  tf.constant(1.0*sample_scaled[:,2],dtype=DTYPE)
    
    x = tf.constant(r,dtype=DTYPE)*tf.cos(theta)
    y = tf.constant(r,dtype=DTYPE)*tf.sin(theta)
    
    print(type(x),x.dtype,type(y),type(t))
    #t = tf.constant(t,dtype=DTYPE)
    #x = tf.constant(x,dtype=DTYPE)
    #y = tf.constant(y,dtype=DTYPE)
    
    return tf.stack((t,x,y),axis=1)
```

```python
def initial_func(p,s = 1.5):
    x = p[:,0:1]
    y = p[:,1:2]
    r = tf.sqrt(tf.square(x)+tf.square(y))
    h = 1.0+0.4*tf.exp(-tf.square(r/s))
    u = tf.zeros_like(h)
    v = tf.zeros_like(h)
    
    print(h.shape,u.shape,v.shape)
    return tf.concat((h,u,v),axis=1)

class domain():
    def __init__(self,lower,upper,DTYPE):
        self.lwb = tf.constant(lower,dtype=DTYPE)
        self.upb = tf.constant(upper,dtype=DTYPE)
        self.dim = tf.constant(len(lower))
        
    def sample_domain(self,n=10):
        x = []
        for i in range(self.dim):
            print(self.lwb[i],self.upb[i])
            x.append(tf.random.uniform((n,1),self.lwb[i],self.upb[i],dtype=DTYPE))
            
        return tf.squeeze(tf.stack(x,axis=1))
    
def get_initial_data(dom,initial_func,n=50,t_ini=0.0):
    t = t_ini*tf.ones((n,1),dtype=DTYPE)
    
    x= [t]
    
    for i in range(1,dom.dim):
        x.append(tf.random.uniform((n,1),dom.lwb[i],dom.upb[i],dtype=DTYPE))
        
    
    
    
    dom_ini = tf.squeeze(tf.stack(x,axis=1))
    f_ini = initial_func(dom_ini[:,1:])
    
    return dom_ini ,f_ini

def get_initial_data2(dom,initial_func,n=50,t_ini=0.0):
    t = t_ini*tf.ones((n,1),dtype=DTYPE)
    
    x= [t]
    
    for i in range(1,dom.dim):
        x.append(tf.random.normal((n,1), mean=0.0,stddev=0.5,dtype=DTYPE))
        
    
    
    
    dom_ini = tf.squeeze(tf.stack(x,axis=1))
    f_ini = initial_func(dom_ini[:,1:])
    
    return dom_ini ,f_ini

def get_bndry_data(dom,n=50):
    
    t = tf.random.uniform((n,1), dom.lwb[0], dom.upb[0], dtype=DTYPE)
    x_rand = tf.random.uniform((n,1), dom.lwb[1], dom.upb[1], dtype=DTYPE)
    y_rand = tf.random.uniform((n,1), dom.lwb[2], dom.upb[2], dtype=DTYPE)
    
    x_low = dom.lwb[1]*tf.ones_like(t)
    x_high = dom.upb[1]*tf.ones_like(t)
    
    y_low = dom.lwb[2]*tf.ones_like(t)
    y_high = dom.upb[2]*tf.ones_like(t)
    
    left = tf.stack((t,x_low,y_rand),axis=1)
    right = tf.stack((t,x_high,y_rand),axis=1)
    
    up = tf.stack((t,x_rand,y_high),axis=1)
    bot = tf.stack((t,x_rand,y_low),axis=1)
    
    x_bndry = tf.concat((left,right,up,bot),axis=0)
    
    f_b0 = tf.ones((x_bndry.shape[0],),dtype=DTYPE)
    f_b1 = tf.zeros((x_bndry.shape[0],),dtype=DTYPE)
    f_b2 = tf.zeros((x_bndry.shape[0],),dtype=DTYPE)
    
    return tf.squeeze(x_bndry),tf.squeeze(tf.stack((f_b0,f_b1,f_b2),axis=1))

def get_bndryInr_data(dom,r_out=6.0,n=50):
    
    t = tf.random.uniform((n,1), dom.lwb[0], dom.upb[0], dtype=DTYPE)
    r = r_out*tf.ones_like(t)
    theta_rand = tf.random.uniform((n,1), 0.0, 2.0*pi, dtype=DTYPE)
    
    x = r*tf.cos(theta_rand)
    y = r*tf.sin(theta_rand)
   
    
    x_bndry = tf.stack((t,x,y),axis=1)
    
    f_b0 = tf.ones_like(x)
    f_b1 = tf.zeros_like(x)
    f_b2 = tf.zeros_like(x)
    
    return tf.squeeze(x_bndry),tf.squeeze(tf.stack((f_b0,f_b1,f_b2),axis=1))

def get_domain_points_lhs(x_min=-1.0,x_max=1.0,y_min=-1.0,y_max=1.0,t_min=0.0,t_max=1.0,n=100):
    
    from scipy.stats import qmc
    sampler = qmc.LatinHypercube(d=3)
    sample = sampler.random(n)
    l_bounds = [t_min,x_min,y_min]
    u_bounds = [t_max,x_max,y_max]
    sample_scaled = qmc.scale(sample, l_bounds, u_bounds)
    t = 1.0*sample_scaled[:,0]
    x = 1.0*sample_scaled[:,1]
    y = 1.0*sample_scaled[:,2]
    
    
    t = tf.constant(t,dtype=DTYPE)
    x = tf.constant(x,dtype=DTYPE)
    y = tf.constant(y,dtype=DTYPE)
    
    return tf.stack((t,x,y),axis=1)
        
```

```python
lower = (0.0,-6.0,-6.0)
upper = (3.0,6.0,6.0)
dom = domain(lower,upper,DTYPE=DTYPE)


#model = get_model(dom,8)

x_i,f_i = get_initial_data(dom,initial_func,n=50)
x_i2,f_i2 = get_initial_data2(dom,initial_func,n=800)
x_b,f_b = get_bndryInr_data(dom,n=500)

print(x_i.shape,x_i.shape,f_i.shape,f_b.shape)

x_bndry = tf.concat((x_i,x_i2,x_b),axis=0)
f_bndry = tf.concat((f_i,f_i2,f_b),axis=0)

x_dom = get_domain_points_lhs(n=3000,x_max=6.0,x_min=-6.0,y_max=6.0,y_min=-6.0,t_max=3.0)
#x_dom = get_domain_pointsInr_lhs(n=10000,t_max=3.0)


#x_dom = dom.sample_domain(4000)
vis = tf.constant(1.0,dtype=DTYPE)

print(x_bndry.shape,f_bndry.shape,x_dom.shape)
r_bndry = tf.sqrt(x_bndry[:,1]**2+x_bndry[:,2]**2)
r_i = tf.sqrt(x_i[:,1]**2+x_i[:,2]**2)
r_i2 = tf.sqrt(x_i2[:,1]**2+x_i2[:,2]**2)
r_b = tf.sqrt(x_b[:,1]**2+x_b[:,2]**2)
#plt.plot(r_i,f_i[:,0],"*")
#plt.plot(x_bndry[:,1],x_bndry[:,2],"*")
plt.plot(x_i[:,1],x_i[:,2],"r*")
plt.plot(x_b[:,1],x_b[:,2],"b*")
plt.plot(x_i2[:,1],x_i2[:,2],"r*")
#plt.plot(r_b)
```

```python
class param_vec_keras():
    def __init__(self,model):
        w = model.get_weights()
        self.lw = len(w)
        self.shapes = []
        self.lens =  []
        for wi in w:
            self.shapes.append(wi.shape)
            self.lens.append(tf.reshape(wi, [-1]).shape[0])
            
        print(len(self.shapes),len(self.lens))
    @tf.function    
    def weights_to_vec(self,wgt):
        print("Tracing weights to vec")
        return tf.concat([tf.reshape(wi, [-1]) for wi in wgt],axis=0)
    @tf.function
    def vec_to_weights(self,wvec):
        print("Tracing vec to weights ")
        wlist=[]
        cum = 0
        for l,s in zip(self.lens,self.shapes):
            ld_wgt = tf.reshape(wvec[cum:cum+l],s)
            wlist.append(ld_wgt)
            cum=cum+l
        
        return wlist

class PINN():
    def __init__(self,dom_bounds,prob_coef,DTYPE='float32'):
        self.DTYPE = DTYPE
        self.lwb = tf.constant(dom_bounds[0],dtype=DTYPE)
        self.upb = tf.constant(dom_bounds[1],dtype=DTYPE)
        self.dim = tf.constant(len(dom_bounds[0]))
        
        self.param_vec: param_vec_keras
        
        
            
        self.prob_coef = tf.constant(prob_coef,dtype=DTYPE)
        
        print("Prob coef",self.prob_coef[0])
        print("Bounds",self.lwb,self.upb)
        
        
        self.model: keras.Model

    def pt(self):
        print(self.model.summary())
        
    def use_vec(self,dom_points,bndry_points,f_bndry):
        self.param_vec = param_vec_keras(self.model)
        
        self.dom_points = dom_points
        self.bndry_points = bndry_points
        self.f_bndry  = f_bndry

        
        
        
    def sample_domain(self,n=10):
        x = []
        for i in range(self.dim):
            print(self.lwb[i],self.upb[i])
            x.append(tf.random.uniform((n,1),self.lwb[i],self.upb[i],dtype=self.DTYPE))
            
        return tf.squeeze(tf.stack(x,axis=1))
    
    
    def make_model(self,no_hid_layers,no_of_neu=20,act='tanh'):
    
        n_inp = tf.constant(self.dim)
        lb = self.lwb
        ub = self.upb
    
        scale_layer = keras.layers.Lambda(lambda x: 2.0*(x - lb)/(ub - lb) - 1.0)
        
        inp = keras.Input(shape=(n_inp,))
        x = scale_layer(inp)
       
      
    
        for i in range(no_hid_layers):
            x = keras.layers.Dense(no_of_neu,activation=tf.keras.activations.get(act),kernel_initializer='glorot_normal')(x)
        
        out = keras.layers.Dense(1)(x)
        
        
    
        return keras.Model(inp,out)
    
    
    def diff_eqn(self,dom_points):
        t,x = dom_points[:,0:1],dom_points[:,1:2]
        
        with tf.GradientTape(persistent=True) as tape:
            tape.watch(t)
            tape.watch(x)
        
            f = self.model(tf.stack((t[:,0],x[:,0]),axis=1))
        
            dx_f = tape.gradient(f,x)
        
        dt_f = tape.gradient(f,t)
        dxx_f = tape.gradient(dx_f,x)
    
        del tape
    
        return dt_f + f*dx_f - self.prob_coef[0]*dxx_f  
    
    def compute_loss(self,dom_points,bndry_points,f_bndry):
        
        diff = self.diff_eqn(dom_points)
        diff_loss = tf.reduce_mean(tf.square(diff))
    
        f_pred = self.model(x_bndry)
        bndry_loss = tf.reduce_mean(tf.square(f_pred-f_bndry))
    
        return diff_loss+bndry_loss
    
    def get_loss_grads_vec_to_vec(self,params_vec):
        
        
        wgt = self.param_vec.vec_to_weights(params_vec)
        
        self.model.set_weights(wgt)
        with tf.GradientTape(persistent=True) as tape:
            tape.watch(self.model.trainable_variables)
        
            loss = self.compute_loss(self.dom_points,self.bndry_points,self.f_bndry)
            #loss = self.compute_loss(dom_points,bndry_points,f_bndry)
        
        grds = tape.gradient(loss,self.model.trainable_variables)
        grds = self.param_vec.weights_to_vec(grds)
        del tape
        
        return (loss,grds) 
    
    def train_step(self,optim,dom_points,bndry_points,f_bndry):
        loss,grds = self.get_grads(dom_points,bndry_points,f_bndry)
        optim.apply_gradients(zip(grds, self.model.trainable_variables))
    
        return loss
    
    def train(self,optim,dom_points,bndry_points,f_bndry,n_epochs = 5000):
        
        losses = []
        t0 = time()
        for i in range(n_epochs):
            loss = self.train_step(optim,dom_points,bndry_points,f_bndry)
            losses.append(losses)
    
            if i%50==0:
                print('It {:05d}: loss = {:10.8e}'.format(i,loss))
        return losses
```
