```julia
using NeuralPDE, Lux, ModelingToolkit, Optimization, OptimizationOptimJL, Roots
using SpecialFunctions
using Plots
import ModelingToolkit: Interval, infimum, supremum
using OrdinaryDiffEq
#using PythonPlot
```

```julia
L=2.0
```

```julia
function ini_u(x)
    ui = -sin(pi*x)
    
    return ui
    
end
```

```julia
hid_units=20
hid_lay_no=8

maxiters=1000

@parameters t, x
@variables u(..)
Dx = Differential(x)
Dxx = Differential(x)^2
Dt = Differential(t)



eqs =Dt(u(t,x)) ~ -u(t,x)*Dx(u(t,x)) - (0.01/pi)*Dxx(u(t,x))

# Nonlinear system of hyperbolic equations
#eqs = [Dtt(u(t, x)) ~ a / (x^n) * Dx(x^n * Dx(u(t, x))) + u(t, x) * f(u(t, x) / w(t, x)),
#    Dtt(w(t, x)) ~ b / (x^n) * Dx(x^n * Dx(w(t, x))) + w(t, x) * g(u(t, x) / w(t, x))]

# Boundary conditions
    
    bcs = [u(0.0,x)~ini_u(x),
            u(t,-1.0) ~ 0.0,
            u(t,1.0) ~ 0.0,
          ]
##bcs = [u(0, x) ~ u_analytic(0, x),
#    w(0, x) ~ w_analytic(0, x),
#    u(t, 0) ~ u_analytic(t, 0),
#    w(t, 0) ~ w_analytic(t, 0),
#    u(t, 1) ~ u_analytic(t, 1),
 #   w(t, 1) ~ w_analytic(t, 1)]

# Space and time domains
    domains = [t ∈ Interval(0.0, 1.0),
    x ∈ Interval(-L*0.5,L*0.5)]

# Neural network
    input_ = length(domains)
    println(input_)
    n = hid_units

    chain = Lux.Chain(Dense(input_, n, Lux.tanh_fast))
        
    for j in range(1,hid_lay_no)
            chain = Lux.Chain(chain,Dense(n,n,Lux.tanh_fast))
            println("chain inner $j")
    end
    chain = Lux.Chain(chain,Dense(n, 1))
    
            
    dx = 0.1
    print(dx)
    strategy = GridTraining(dx)
    discretization = PhysicsInformedNN(chain, strategy)
        
        
        

    @named pdesystem = PDESystem(eqs, bcs, domains, [t, x], [u(t, x)])
    prob = discretize(pdesystem, discretization)
    sym_prob = symbolic_discretize(pdesystem, discretization)

    pde_inner_loss_functions = sym_prob.loss_functions.pde_loss_functions
    bcs_inner_loss_functions = sym_prob.loss_functions.bc_loss_functions

    callback = function (p, l)
     println("loss: ", l)
     println("pde_losses: ", map(l_ -> l_(p), pde_inner_loss_functions))
     println("bcs_losses: ", map(l_ -> l_(p), bcs_inner_loss_functions))
     return false
    end

    
   

```

```julia
dxs = [dx,dx]
a=[infimum(d.domain):dx:supremum(d.domain) for (d, dx) in zip(domains, dxs)]
```

```julia
collect.(a)
```

```julia
res = Optimization.solve(prob, BFGS(),callback=callback; maxiters = 300)
```
